```
# -*- coding: utf-8 -*-
from chainer import Chain
import chainer.links as L
import chainer.functions as F
# 数値計算関連
import math
import tensorflow as tf
import random
import numpy as np
import matplotlib.pyplot as plt
# chainer
from chainer import Chain, Variable
import chainer.functions as F
import chainer.links as L
from chainer import optimizers

class MyChain(Chain):

    def __init__(self):
        super(MyChain, self).__init__(
            l1 = L.Linear(1, 100),
            l2 = L.Linear(100, 30),
            l3 = L.Linear(30, 1)
        )

    def predict(self, x):
        h1 = F.relu(self.l1(x))
        h2 = F.relu(self.l2(h1))
        return self.l3(h2)
noise=0.05*np.random.rand(100).astype(np.float32) #ノイズを作成


# 乱数のシードを固定
random.seed(1)

# 標本データの生成
#   真の関数としてsin関数を
x, y = [], []
x_data=np.random.rand(100).astype(np.float32)  #ランダムな位置を作成(小数)  0~1までの乱数を100こ作る
noise=0.05*np.random.rand(100).astype(np.float32) #ノイズを作成
theta=30 #θ
for i in np.linspace(-3,3,100):
    x.append([i*x_data])
    y.append([math.sin(i)])  # 真の関数
    #x.append([x_data])
    #y_data=np.sin(2*np.pi*x_data**2)+theta+noise #観測用データ
    

# NNモデルを宣言
model = MyChain()

# chainerの変数として再度宣言
x = Variable(np.array(x, dtype=np.float32))
y = Variable(np.array(y, dtype=np.float32))



# 損失関数の計算
#   損失関数には自乗誤差(MSE)を使用
def forward(x, y, model):
    t = model.predict(x)
    loss = F.mean_squared_error(t, y)
    return loss

# chainerのoptimizer
#   最適化のアルゴリズムには Adam を使用
optimizer = optimizers.Adam()
# modelのパラメータをoptimizerに渡す
optimizer.setup(model)

# パラメータの学習を繰り返す
for i in range(0,1000):
    loss = forward(x, y, model)
    print(loss.data)  # 現状のMSEを表示
    optimizer.update(forward, x, y, model)

# プロット
t = model.predict(x)
plt.plot(x.data, y.data)
plt.scatter(x.data, t.data)
plt.grid(which='major',color='gray',linestyle='-')
plt.ylim(-1.5, 1.5)
plt.xlim(-4, 4)
plt.show()        
```
